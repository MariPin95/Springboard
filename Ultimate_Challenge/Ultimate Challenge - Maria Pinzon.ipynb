### Data preparation and cleaning

```python
with open('./ultimate_data_challenge.json') as f:
    data = json.load(f)

ultimate_df = pd.DataFrame(data)
ultimate_df.head()
```

```python
ultimate_df.info()
```

```python
ultimate_df.describe()
```

```python
ultimate_df.isna().sum()
```

```python
ultimate_df.dropna(subset = 'phone', inplace = True)
```

```python
ultimate_df['avg_rating_of_driver'].describe()
```

```python
ultimate_df['avg_rating_of_driver'].mean()
```

```python
# fill na with mean
mean = ultimate_df['avg_rating_of_driver'].mean()
ultimate_df['avg_rating_of_driver'] = ultimate_df['avg_rating_of_driver'].fillna(mean)
```

```python
ultimate_df['avg_rating_by_driver'].describe()
```

```python
# fill na with average of mean and median
fill_val = (ultimate_df['avg_rating_by_driver'].mean() + 5) / 2
ultimate_df['avg_rating_by_driver'] = ultimate_df['avg_rating_by_driver'].fillna(fill_val)
```

```python
ultimate_df.info()
```

```python
# convert to datetime
ultimate_df[['signup_date', 'last_trip_date']] = ultimate_df[['signup_date','last_trip_date']].apply(pd.to_datetime, format='%Y-%m-%d')
```

```python
latest_date = ultimate_df['last_trip_date'].max()
```

```python
# create column for active users
ultimate_df['active'] = ultimate_df['last_trip_date'].apply(lambda date: 0 if (latest_date-date).days>30 else 1)
ultimate_df.head()
```

```python
# Percentage of active users
ultimate_df['active'].value_counts()
```

### Feature Engineering

```python
df_features = pd.get_dummies(ultimate_df, columns = ['city', 'phone', 'ultimate_black_user'])
df_features.head()
```

```python
target = ultimate_df['active']
```

```python
# drop unused columns from feature dataset
df_features = df_features.drop(['last_trip_date','signup_date', 'active'], axis=1)
```

### Modeling

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
```

```python
ultimate_df.active.value_counts(normalize=True)*100
```

```python
scaler = QuantileTransformer()
scaled_features = scaler.fit_transform(df_features)
```

```python
# set X and Y
X = df_features
Y = target
```

```python
#split data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=42)
```

```python
# First model built
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
rf.score(X_test,y_test)
```

```python
print(cross_val_score(rf,X_test,np.ravel(y_test),cv=5))
print('Mean Cross Validated Score: ', np.mean(cross_val_score(rf,X_test,np.ravel(y_test),cv=5)))
```

```python
importance_features = pd.DataFrame(zip(list(df_features.columns),rf.feature_importances_),index=range(df_features.columns.shape[0]),columns=['feature','importance'])
importance_features.sort_values(by='importance',ascending=False)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# Second model
lf_grid = LogisticRegression()
param_grid = {'C':np.arange(0.5,5.1,0.1)}
grid = GridSearchCV(lf_grid,param_grid,cv=5)
grid.fit(X_train,np.ravel(y_train))
```

```python
grid.best_estimator_.C
```

```python
lf_coef = LogisticRegression(C=grid.best_estimator_.C)
lf_coef.fit(X_train,y_train)
y_pred = lf_coef.predict(X_test)
```

```python
lf_coef.score(X_test,y_test)
```

```python
importance_features_lf = pd.DataFrame(zip(list(df_features.columns),lf_coef.coef_[0]),index=range(df_features.columns.shape[0]),columns=['feature','importance'])
importance_features_lf.sort_values(by='importance',ascending=False)
```

Analyzing both models, Random Forest does slightly better at predicting than Logistic Regression. The model has approximately a 75% accuracy score, which could be improved but good information can be learnt from it. 

Specifically, it shows us the features that are most important to analyze: 

    Average distance = This shows that active users tend to have longer active distance than inactive.
    
    Weekday_pct = This shows that activity on the weekends is a good indicator of active users which can tell the company to incentivize users by giving promotions on the weekends.
    
    Avg_rating_by_drive = This seems to be pretty equal from active to non-active users, which does not really give an insight on how to retain user.
    

Based on this, it seems that the company should be focusing on promotions during the weekends as it seems to be the most important factor
